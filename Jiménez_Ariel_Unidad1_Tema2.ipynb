## Parte 1: Fase de Limpieza
### 1. Importación de Librerías

```python
# Importamos dataclass, un decorador que nos ahorra escribir código repetitivo 
from dataclasses import dataclass
# Importamos 're', el módulo de Python para trabajar con Expresiones Regulares.
import re
```

### 2. Carga del Dataset Crudo

```python
# Dataset de científicos con información desordenada
data_set = [ 
    "ID:001; NAME:  alberto_EINSTEIN  ; PROYECTO:Relatividad; HORAS:40; PAGO_H:35.5", 
    "id-002, name- marie_CURIE, proyecto-Radiactividad, horas-35, pago_h-40.0", 
    "ID:003; NAME:isaac_NEWTON; PROYECTO:Leyes del Movimiento; HORAS:20; PAGO_H:30.0####", 
    "   id:004 - name: nikola_TESLA - proyecto:Corriente Alterna - horas:50 - pago_h:25.0", 
    "ID:005; NAME:ada_LOVELACE; PROYECTO:Algoritmos; HORAS:45; PAGO_H:45.0", 
    "id-006, name- charles_DARWIN, proyecto-Evolución, horas-30, pago_h-20.0????", 
    "ID:007; NAME:  rosalind_FRANKLIN  ; PROYECTO:Estructura ADN; HORAS:38; PAGO_H:32.5", 
    "id:008 - name: stephen_HAWKING - proyecto:Agujeros Negros - horas:15 - pago_h:50.0", 
    "ID:009; NAME:katherine_JOHNSON; PROYECTO:Trayectorias NASA; HORAS:42; PAGO_H:40.0!!!!", 
    "id-010, name- grace_HOPPER, proyecto-Compiladores, horas-48, pago_h-38.0" 
]
# Se imprime en la consola el primer elemento de la lista data_set para verificar visualmente que la información cruda fue cargada correctamente en memoria.
print(data_set[0])
```

### 3. Definición de la Estructura de Datos (Dataclass)

```python
# Se aplica el decorador @dataclass a la clase subsecuente para simplificar su declaración y manejar automáticamente el almacenamiento del estado.
@dataclass
# Se declara la clase Investigador, la cual actúa como una estructura de datos estandarizada para almacenar la información limpia de cada científico.
class Investigador:
    # Se define el atributo id, el cual almacenará el identificador numérico único del investigador bajo un tipo de dato entero.
    id: int
    # Se define el atributo investigador, el cual contendrá el nombre formateado del científico utilizando un tipo de dato de cadena de texto.
    investigador: str
    # Se define el atributo area, el cual guardará el nombre del proyecto de investigación en un tipo de dato de cadena de texto.
    area: str
    # Se define el atributo salario_semanal, destinado a almacenar el resultado financiero calculado bajo un tipo de dato de punto flotante.
    salario_semanal: float
```

### 4. Función de Normalización y Limpieza
*Manejo de separadores, formato de texto, tipado y cálculo de salario.*

```python
# Se define la función normalizar, la cual recibe como parámetro raw_data correspondiente a una cadena de texto sucia que necesita procesamiento.
def normalizar(raw_data):
    # Se eliminan los espacios en blanco sobrantes al inicio y al final de la cadena de texto original mediante el método strip.
    linea = raw_data.strip() 

    # Se emplea una expresión regular para sustituir múltiples caracteres separadores inconsistentes (comas, punto y coma, barras, guiones, dos puntos) por una coma estándar.
    linea_normalizada = re.sub(r"[,;|\-_:]", ",", linea)

    # Se divide la cadena normalizada utilizando la coma como delimitador y se aplica una comprensión de listas para limpiar espacios y capitalizar cada fragmento resultante.
    cadena_final = [l.strip().capitalize() for l in linea_normalizada.split(",")]
    
    # Se extrae el segundo elemento de la lista procesada (índice 1), se convierte a un número entero y se asigna a la variable id.
    id = int(cadena_final[1])

    # Se extraen los elementos correspondientes al nombre y apellido (índices 3 y 4), uniéndolos con un espacio en blanco para formar el nombre completo.
    nombre = " ".join(cadena_final[3:5])

    # Se captura el nombre del proyecto en el índice 6 y se convierte a letras mayúsculas para estandarizar su presentación.
    proyecto = cadena_final[6].upper()

    # Se obtiene el dato de las horas trabajadas en el índice 8, transformándolo a un valor numérico de tipo entero.
    horas = int(cadena_final[8])

    # Se extrae el último elemento de la lista procesada, el cual corresponde al pago por hora, y se almacena en una variable temporal por contener posibles caracteres residuales.
    temp = cadena_final[-1]

    # Se define una lista auxiliar con los símbolos de puntuación no deseados que se han identificado al final de algunos montos de pago.
    separadores_pago = ["!", "?", "#"]

    # Se inicia un ciclo iterativo que recorre cada uno de los caracteres no deseados definidos previamente.
    for s in separadores_pago: 
        # Se reemplaza el símbolo iterado por una cadena vacía en la variable temporal, eliminándolo efectivamente del texto.
        temp = temp.replace(s, "") 

    # Se limpia nuevamente la variable temporal de posibles espacios residuales generados tras las eliminaciones.
    pago_h = temp.strip() 

    # Se convierte la cadena de texto limpia del pago por hora a un tipo de dato de punto flotante para posibilitar operaciones aritméticas.
    pago_h = float(pago_h) 
    
    # Se calcula el salario semanal total multiplicando las horas trabajadas por la tarifa de pago por hora.
    salario_semanal = horas * pago_h 

    # Se retorna una tupla estructurada con los datos procesados y listos: el identificador, nombre, proyecto, horas y el salario semanal.
    return id, nombre, proyecto, horas, salario_semanal
```

### 5. Procesamiento y Extracción de Registros
*Ejecución de la limpieza y almacenamiento en estructuras de datos.*

```python
# Se inicializa una lista vacía denominada horas_registro, cuyo propósito es acumular de forma independiente las horas trabajadas por todos los científicos.
horas_registro = []

# Se inicializa una lista vacía llamada lista_cientificos, diseñada para almacenar los objetos estructurados de la clase Investigador que se instanciarán posteriormente.
lista_cientificos = []
```

```python
# Se establece un ciclo iterativo para procesar una a una cada cadena de texto contenida en el conjunto de datos data_set.
for linea in data_set:
    # Se ejecuta la función normalizar sobre la línea actual, desempaquetando los cinco valores retornados en sus variables individuales correspondientes.
    id, nombre, proyecto, horas, salario_semanal = normalizar(linea)

    # Se añade la cantidad de horas extraídas en la iteración actual al registro acumulativo de horas globales.
    horas_registro.append(horas) 

    # Se crea una nueva instancia de la clase Investigador, proveyendo los datos normalizados para poblar los atributos del objeto.
    cientifico = Investigador(id, nombre, proyecto, salario_semanal) 

    # Se adjunta el objeto estructurado del científico a la colección principal lista_cientificos.
    lista_cientificos.append(cientifico)
```

## Parte 3: Personalización de la Tarea

### 1. Función: Ordenamiento Dinámico y Generación de Códigos
*Función propia para clasificar datos bajo múltiples criterios y generar IDs únicos.*

```python
# Se define la función reporte, la cual parametriza la lista de investigadores, el criterio de ordenación, la dirección (ascendente/descendente) y un límite opcional de resultados.
def reporte(lista_cient, criterio="salario", orden="desc", top_num=None):
    # Se crea una copia superficial de la lista original de científicos para evitar mutar los datos base al momento de aplicar el algoritmo de ordenamiento.
    lista = lista_cient.copy()

    # Se obtiene y almacena la cantidad total de elementos dentro de la lista copiada para gestionar los ciclos del algoritmo.
    n = len(lista)

    # Se inicia el bucle externo del algoritmo de ordenamiento burbuja que iterará n veces sobre el arreglo.
    for i in range(n):
        # Se inicia el bucle interno que compara los elementos adyacentes; el límite se reduce en i para no 
        # evaluar elementos ya ordenados al final del arreglo.
        for j in range(0, n - i - 1):
            # Se evalúa si el criterio de ordenamiento solicitado por el usuario es en base al salario.
            if  criterio == "salario":
                # Se asigna el salario del elemento en la posición actual a una variable de valor comparativo primario.
                val1 = lista[j].salario_semanal
                # Se asigna el salario del elemento en la posición adyacente a una variable de valor comparativo secundario.
                val2 = lista[j+1].salario_semanal           
            # Alternativamente, se evalúa si el criterio de ordenamiento corresponde a la longitud del proyecto.
            elif criterio == "proyecto":
                # Se determina la cantidad de caracteres del área/proyecto en la posición actual como valor comparativo primario.
                val1 = len(lista[j].area)
                # Se determina la cantidad de caracteres del área/proyecto de la posición adyacente como valor comparativo secundario.
                val2 = len(lista[j+1].area) 
            
            # Se inicializa una bandera booleana en falso, asumiendo inicialmente que no se requiere intercambiar posiciones en este paso.
            intercambio = False

            # Se verifica si el orden requerido es ascendente y el valor primario supera al secundario, lo que obligaría a un cambio.
            if orden == "asc" and val1 > val2:
                # Se activa la bandera para confirmar que los elementos deben permutar de lugar.
                intercambio = True
            # Se verifica la condición contraria: si el orden es descendente y el valor primario es inferior al secundario.
            elif orden == "desc" and val1 < val2:
                # Se activa la bandera para proceder con el intercambio posicional.
                intercambio = True
            
            # Se evalúa el estado final de la bandera para esta iteración.
            if intercambio:
                # Se ejecuta el intercambio directo de posiciones entre el objeto actual y el adyacente dentro de la lista.
                lista[j], lista[j+1] = lista[j+1], lista[j] 

    # Se evalúa si el parámetro para limitar el número superior de resultados contiene un valor válido y no es nulo.
    if top_num is not None:
        # Se aplica un slicing a la lista para conservar únicamente el número de registros determinado por el parámetro top_num.
        lista = lista[:top_num]
            
    # Se imprime un encabezado dinámico que notifica el criterio y la dirección que se emplearon para efectuar el ordenamiento de la lista mostrada.
    print(f"Lista ordenada según '{criterio}' de forma '{orden}'")

    # Se imprime la cabecera formateada de la tabla para visualizar los resultados ordenados, destacando una nueva columna para un identificador dinámico.
    print(f"{'ID Nuevo':<15} | {'Nombre':<20} | {'Proyecto':<25} | {'Salario Semanal':<15}")

    # Se inicia un ciclo para recorrer la lista de investigadores que ya se encuentra ordenada y opcionalmente recortada.
    for cient in lista:

        # Se construye un nuevo código concatenando las tres primeras letras del nombre y las tres últimas del área, transformando el resultado final a mayúsculas.
        codigo_completo = (cient.investigador[:3] + cient.area[-3:]).upper()

        # Se crea el formato definitivo del identificador nuevo, uniendo el código alfabético construido con el identificador numérico real del científico.
        cod_fmt = f"{codigo_completo}-{cient.id}"

        # Se imprime la fila del registro del investigador con el ancho estipulado para cada columna, mostrando el nuevo ID y manteniendo el formato monetario.
        print(f"{cod_fmt:<15} | {cient.investigador:<20} | {cient.area:<25} | ${cient.salario_semanal:<15.2f}")
    
    # La función finaliza retornando la lista ya procesada, permitiendo su almacenamiento o reuso externo.
    return lista
```
