```python
# Define la estructura de un nodo para una lista circular doblemente enlazada de monitoreo.
class  MonitoreoPaciente:
    # Inicializa las propiedades biomédicas y de identificación del paciente en su cama.
    def __init__(self, id_cama: int, nombre_paciente: str, frecuencia_cardiaca: int, estado_alerta: str):
        # Asigna el número identificador de la cama asignada.
        self.id_cama = id_cama     
        # Guarda el nombre completo del paciente ingresado.
        self.nombre_paciente = nombre_paciente   
        # Registra la medición inicial de la frecuencia cardíaca.
        self.frecuencia_cardiaca = frecuencia_cardiaca
        # Establece la condición médica inicial del paciente.
        self.estado_alerta = estado_alerta
        # Inicializa el puntero de enlace posterior en nulo.
        self.sig = None          
        # Inicializa el puntero de enlace anterior en nulo.
        self.ant = None             

# Define el controlador para administrar los pacientes en la Unidad de Cuidados Intensivos.
class PacientesUCI:
    # Constructor que inicializa el sistema de monitoreo circular sin registros.
    def __init__(self):
        # Establece el primer nodo de referencia de la sala como vacío.
        self.cabeza = None
        # Inicializa el marcador de lectura del monitor actual en nulo.
        self.monitor_actual = None
    
    # Comprueba si la sala médica carece de pacientes conectados.
    def esta_vacio(self):
        # Retorna una condición lógica basada en el estado de la cabeza.
        return self.cabeza is None

    # Localiza un nodo mediante comparación de texto o identificación numérica.
    def buscar_paciente(self, valor, criterio="nombre"):
        # Verifica el estado de la lista antes de comenzar la inspección cíclica.
        if self.cabeza is None:
            # Retorna un valor nulo debido a la ausencia de elementos.
            return None
        # Establece el punto de inicio para la búsqueda en la cabeza.
        actual = self.cabeza
        # Sostiene el proceso de validación sobre el anillo de nodos.
        while True:
            # Evalúa la coincidencia del nombre transformando ambos elementos a minúsculas.
            if criterio == "nombre" and actual.nombre_paciente.lower() == str(valor).lower():
                # Devuelve el objeto del nodo que satisface la condición textual.
                return actual
            # Comprueba si el criterio busca coincidencia exacta con la cama asignada.
            elif criterio == "id" and actual.id_cama == valor:
                # Entrega el objeto del nodo que coincide con el número entero.
                return actual
            # Desplaza el rastreo hacia el nodo continuo a la derecha.
            actual = actual.sig
            # Determina si el ciclo dio una vuelta completa sin hallar resultados.
            if actual == self.cabeza: 
                # Concluye la búsqueda retornando un valor nulo.
                return None
            
    # Integra un nuevo paciente cerrando la estructura de forma circular por ambos extremos.
    def conectar_paciente(self, id_cama, nombre_paciente, frecuencia_cardiaca, estado_alerta):
        # Llama a la función de búsqueda para verificar si la cama ya está ocupada por otro paciente.
        cama_ocupada = self.buscar_paciente(id_cama, "id")
        # Si la función de búsqueda retorna un nodo, significa que la cama ya está ocupada y no se puede asignar a otro paciente.
        if cama_ocupada is not None:
            print(f"La cama {id_cama} ya está ocupada por el paciente {cama_ocupada.nombre_paciente}. No se puede conectar a {nombre_paciente}.")
            # Detiene la ejecución de la función para evitar la asignación de un paciente a una cama ya ocupada.
            return
        # Instancia un nodo de la clase MonitoreoPaciente con las variables provistas.
        nuevo_paciente =  MonitoreoPaciente(id_cama, nombre_paciente, frecuencia_cardiaca, estado_alerta)

        print(f"Paciente {nombre_paciente} conectado a la cama {id_cama}.")
        
        # Evalúa si la estructura de monitoreo se encuentra desocupada.
        if self.cabeza is None:
            # Convierte al nuevo paciente en el nodo raíz de la sala.
            self.cabeza = nuevo_paciente 
            # Direcciona el enlace posterior hacia el propio nodo para iniciar el círculo.
            self.cabeza.sig = self.cabeza
            # Direcciona el enlace anterior hacia sí mismo para cerrar la doble vía.
            self.cabeza.ant = self.cabeza
            # Posiciona la lectura del monitor en este primer registro disponible.
            self.monitor_actual = self.cabeza
        # Flujo de ejecución cuando la lista circular ya posee elementos.
        else:
            # Inicializa un puntero auxiliar situado en la cabeza del ciclo.
            paciente_actual = self.cabeza 
            
            # Recorre los nodos hasta hallar el que se enlaza de vuelta con la cabeza.
            while paciente_actual.sig != self.cabeza: 
                # Transfiere la exploración hacia el nodo derecho adyacente.
                paciente_actual = paciente_actual.sig 
                
            # Conecta la salida del antiguo último paciente con el nuevo nodo.
            paciente_actual.sig = nuevo_paciente 
            # Conecta la entrada previa del nuevo nodo con el penúltimo paciente.
            nuevo_paciente.ant = paciente_actual
            
            # Apunta la salida del nuevo nodo terminal de vuelta a la cabeza.
            nuevo_paciente.sig = self.cabeza 
            # Redirige el brazo anterior de la cabeza hacia el nuevo nodo terminal.
            self.cabeza.ant = nuevo_paciente
    
    # Desplaza de forma interactiva el monitor actual hacia adelante o hacia atrás.
    def navegacion_pacientes(self, forma):
        # Evalúa si la instrucción recibida demanda avanzar en la secuencia circular.
        if forma in ["sig", "siguiente", "s"]:
            print(f"Paciente {self.monitor_actual.sig.nombre_paciente}")
            # Avanza la posición activa del monitor al siguiente nodo de la derecha.
            self.monitor_actual = self.monitor_actual.sig
            
        # Comprueba si el parámetro solicita retroceder en el ciclo cerrado.
        elif forma in ["ant", "anterior", "a"]:
            print(f"Paciente {self.monitor_actual.ant.nombre_paciente}")
            # Retrocede la posición activa del monitor al nodo de la izquierda.
            self.monitor_actual = self.monitor_actual.ant

    # Simula la actualización aleatoria y análisis de las constantes vitales.
    def analisis_fc(self):
        # Valida la existencia de datos para prevenir fallos en el ciclo.
        if self.cabeza is None:
            print("No hay pacientes conectados.")
            # Devuelve un valor nulo para indicar la ausencia de procesamiento.
            return None
        
        # Define el inicio del barrido en el nodo principal de la UCI.
        actual = self.cabeza
        print("Iniciando análisis de frecuencia cardíaca para todos los pacientes\n")
        # Inicia un ciclo de procesamiento infinito controlado por retorno de base.
        while True:
            # Genera y asigna un pulso cardíaco aleatorio al paciente evaluado.
            actual.frecuencia_cardiaca = random.randint(40, 140)

            # Evalúa si la constante vital se encuentra fuera del rango saludable.
            if actual.frecuencia_cardiaca < 60 or actual.frecuencia_cardiaca > 120:
                # Modifica el diagnóstico del nodo a una condición de riesgo.
                actual.estado_alerta = "Crítico"
                print(f"¡ALERTA! Paciente {actual.nombre_paciente} en estado {actual.estado_alerta} - FC: {actual.frecuencia_cardiaca}")
            # Ejecución en caso de que la constante vital sea fisiológicamente normal.
            else:
                # Restablece el diagnóstico del nodo a una condición segura.
                actual.estado_alerta = "Estable"
                print(f"Paciente {actual.nombre_paciente} en estado {actual.estado_alerta} - FC: {actual.frecuencia_cardiaca}")
            # Mueve el cursor de análisis al siguiente paciente del anillo.
            actual = actual.sig

            # Rompe el bucle al comprobar que el cursor regresó al punto de partida.
            if actual == self.cabeza:
                print("\nAnálisis de frecuencia cardíaca completado para todos los pacientes") 
                # Concluye la iteración tras completar una vuelta al ciclo.
                break

    # Remueve un registro del ciclo cerrado reestructurando los enlaces contiguos.
    def desconectar_paciente(self, id_cama):
        # Valida que la estructura tenga registros previo a realizar la baja.
        if self.esta_vacio():
            # Devuelve falso al no poder efectuar ninguna remoción de datos.
            return False
        # Solicita la localización del nodo específico mediante su identificador.
        dado_alta = self.buscar_paciente(id_cama, "id")

        # Comprueba si el proceso de búsqueda no detectó al paciente indicado.
        if dado_alta is None:
            # Retorna una confirmación negativa por ausencia del registro.
            return False
        
        # Evalúa si el paciente es el único nodo remanente en el ciclo cerrado.
        if dado_alta.sig == dado_alta:
            # Remueve la referencia inicial vaciando el sistema.
            self.cabeza = None
            # Limpia la posición del monitor activo del sistema.
            self.monitor_actual = None
        # Bloque de reestructuración si existen múltiples nodos enlazados.
        else:
            # Extrae el enlace al nodo vecino ubicado a la izquierda.
            nodo_ant = dado_alta.ant
            # Extrae el enlace al nodo vecino ubicado a la derecha.
            nodo_sig = dado_alta.sig

            # Enlaza la salida del vecino izquierdo directamente con el derecho.
            nodo_ant.sig = nodo_sig
            # Modifica el retorno del vecino derecho directamente hacia el izquierdo.
            nodo_sig.ant = nodo_ant

            # Evalúa si el elemento que se elimina es la cabeza actual.
            if dado_alta == self.cabeza:
                # Traslada la posición de la cabeza al nodo adyacente derecho.
                self.cabeza = dado_alta.sig
            
            # Evalúa si el elemento que se elimina sostiene el monitor activo.
            if dado_alta == self.monitor_actual:
                # Desplaza el marcador del monitor hacia el nodo adyacente derecho.
                self.monitor_actual = dado_alta.sig

        # Aísla la salida del nodo desvinculado asignándole un valor nulo.
        dado_alta.sig = None
        # Aísla la entrada del nodo desvinculado asignándole un valor nulo.
        dado_alta.ant = None
        # Concluye con éxito confirmando la remoción del registro.
        return True

    # Realiza un recorrido completo del anillo imprimiendo sus propiedades.
    def mostrar_lista(self):
        if self.esta_vacio():
            print("No hay pacientes conectados.")
            # Termina la función sin realizar ningún proceso de impresión.
            return
        # Inicializa el puntero de lectura en el nodo raíz de la UCI.
        actual = self.cabeza

        # Establece un contador entero para registrar el orden físico.
        orden = 1
        # Mantiene la visualización activa mientras el puntero sea válido.
        print("Lista de pacientes conectados:")
        while True:
            print(f"Paciente: {actual.nombre_paciente} - ID Cama: {actual.id_cama} - Frecuencia Cardiaca: {actual.frecuencia_cardiaca} - Estado de Alerta: {actual.estado_alerta} - Posición en lista: {orden}")
            # Pasa el control de impresión al nodo subsiguiente.
            actual = actual.sig
            # Incrementa en una unidad el valor numérico de la posición.
            orden += 1
            # Verifica si el puntero cerró la lectura al retornar al origen.
            if actual == self.cabeza: 
                # Interrumpe el ciclo de impresión para evitar un bucle sin fin.
                break
```

```python
# Función que despliega el menú de opciones para la gestión de pacientes en la UCI.
def menu():
    print("\n UCI - Monitoreo de Pacientes")
    print("1. Conectar paciente")
    print("2. Buscar paciente por nombre")
    print("3. Navegación ant/sig pacientes")
    print("4. Analizar frecuencia cardiaca de pacientes")
    print("5. Dar de alta paciente")
    print("6. Mostrar lista de pacientes")
    print("7. Salir")

# Función que solicita la entrada de datos para conectar un nuevo paciente a la UCI.
def entrada_datos(paciente):
    # Inicia un ciclo de validación para asegurar la integridad de los datos ingresados.
    while True:
        try:
            # Ingreso de id de cama, nombre del paciente, frecuencia cardíaca y estado de alerta.
            id = int(input("ID cama: "))
            n = input("Nombre paciente: ").strip()
            fc = int(input("Frecuencia cardiaca: "))
            a = input("Estado de alerta: ").strip()
            if id > 0 and fc > 0 and fc < 200 and a and n:
                paciente.conectar_paciente(id, n, fc, a)
                break
            else:
                print("Datos no válidos. Intente de nuevo.")
        except ValueError:
            print("Asegúrese de ingresar correctamente los datos.")
```

```python
# Define la función principal que administra el entorno del menú interactivo de la UCI.
def interfaz():
    # Inicializa una instancia limpia de la clase PacientesUCI para el control de la sala.
    paciente = PacientesUCI() 
    # Sostiene un bucle iterativo para mantener la disponibilidad continua de las opciones.
    while True:
        # Invoca a la función encargada de plasmar el texto del panel en pantalla.
        menu()
        # Captura la opción digitada por el usuario para redirigir el flujo.
        opcion = input("Seleccione una opción: ")

        # Evalúa si la intención del operador consiste en ingresar un nuevo paciente.
        if opcion == "1":
            # Llama al asistente externo de entrada de datos pasando la estructura de control.
            entrada_datos(paciente)

        # Comprueba si la selección equivale a la búsqueda nominal de un registro.
        elif opcion == "2":
            # Solicita el criterio de texto para realizar el rastreo en el anillo.
            p = input("Nombre paciente a buscar: ")
            # Ejecuta la búsqueda parametrizada y guarda el objeto o nulo resultante.
            a_buscar = paciente.buscar_paciente(p, "nombre")
            
            # Valida si el puntero retornado por el buscador es equivalente a vacío.
            if a_buscar is None:
                print("No se encontró el paciente o la lista está vacía.")
            # Ejecuta la lectura de atributos si el nodo existe en la secuencia circular.
            else:
                print (f"Paciente encontrado: {a_buscar.nombre_paciente} - ID Cama: {a_buscar.id_cama} - Frecuencia Cardiaca: {a_buscar.frecuencia_cardiaca} - Estado de Alerta: {a_buscar.estado_alerta}")
        
        # Determina si el comando ingresado solicita activar la inspección de nodos adyacentes.
        elif opcion == "3":
            # Inicializa una variable de texto vacía para la regla de dirección.
            forma = ""
            
            # Verifica si la estructura de monitoreo carece de registros para navegar.
            if paciente.esta_vacio():
                print("No hay pacientes conectados.")
            # Si el sistema posee registros, se activa el ciclo de navegación de pacientes.
            else:
                print(f"Empezando navegación desde el paciente: {paciente.monitor_actual.nombre_paciente }")
                # Abre un ciclo interno dedicado exclusivamente al desplazamiento por el anillo.
                while True:
                    # Captura la instrucción de dirección o el comando para romper el bucle.
                    forma = input("¿Desea ver el siguiente paciente (s) o el anterior (a)?, n para salir: ").lower().strip()
                    # Evalúa si la instrucción de dirección es equivalente a la orden de salida del ciclo.
                    if forma == "n":
                        # Destruye el ciclo interno de navegación para retornar al menú principal.
                        break
                    # Llama al método de navegación pasando la dirección indicada por el operador.
                    paciente.navegacion_pacientes(forma)
                    
        # Evalúa si la opción elegida requiere procesar las constantes vitales aleatorias.
        elif opcion == "4":
            # Activa el algoritmo que modifica y evalúa los ritmos cardíacos de los pacientes.
            paciente.analisis_fc()

        # Comprueba si se seleccionó tramitar la baja médica de una cama.
        elif opcion == "5":
            # Inicializa un marcador entero temporal para almacenar el identificador de la cama.
            p = 0
            # Sostiene un lazo de validación hasta garantizar un ingreso numérico correcto.
            while True:
                # Inicia un entorno de contención contra errores de conversión tipográfica.
                try: 
                    # Transforma la entrada de consola directamente a formato numérico entero.
                    p = int(input("Ingrese ID cama paciente: "))
                    # Valida que el identificador posea una magnitud entera positiva.
                    if p > 0:
                        # Rompe la secuencia de captura al cumplir con la regla aritmética.
                        break
                # Atrapa la anomalía de ejecución en caso de recibir caracteres alfabéticos.
                except ValueError:
                    print("Ingresar un número entero.")
            # Ejecuta el método de desincorporación y almacena su confirmación lógica.
            dado_alta = paciente.desconectar_paciente(p)
            # Evalúa si la reestructuración de los punteros circulares fue exitosa.
            if dado_alta:
                print(f"Paciente con ID de cama {p} dado de alta.")
            # Acción secundaria si el identificador numérico no fue hallado en el anillo.
            else:
                print("No se encontró el paciente para dar de alta o la lista está vacía.")

        # Valida si la opción digitada solicita la impresión de la sala entera.
        elif opcion == "6":
            # Llama de forma directa al método de mapeo secuencial cíclico.
            paciente.mostrar_lista()
        
        # Determina si el operador solicita finalizar por completo el script.
        elif opcion == "7":
            print("Saliendo de la plataforma") 
            # Rompe el bucle maestro interrumpiendo el funcionamiento del sistema.
            break
        
        # Captura cualquier opción del menú que burle las alternativas numeradas.
        else:
            print("Opción inválida.")
# Ejecuta la función principal para iniciar el sistema de gestión de pacientes en la UCI.
interfaz()
```
