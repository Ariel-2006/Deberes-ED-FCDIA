```python
# Define la estructura de un nodo para una lista doblemente enlazada de pacientes.
class PacienteEspera:
    # Inicializa las propiedades fundamentales de cada paciente en la lista.
    def __init__(self, numero_turno: int, nombre_paciente: str, hora_llegada: int, motivo_consulta: str):
        # Asigna el número identificador de turno del paciente.
        self.numero_turno = numero_turno
        # Almacena el nombre completo del paciente.
        self.nombre_paciente = nombre_paciente     
        # Registra el valor numérico de la hora de arribo.
        self.hora_llegada = hora_llegada
        # Guarda la descripción médica de la consulta.
        self.motivo_consulta = motivo_consulta 
        # Inicializa el puntero de enlace hacia adelante en nulo.
        self.sig = None          
        # Inicializa el puntero de enlace hacia atrás en nulo.
        self.ant = None         

# Define la estructura de control para administrar la cola doblemente enlazada.
class SalaEspera:
    # Constructor que inicializa la sala de espera sin ningún elemento.
    def __init__(self):
        # Establece el primer nodo de la fila como vacío.
        self.cabeza = None   
    
    # Determina la presencia o ausencia de pacientes en la estructura.
    def esta_vacia(self):
        # Retorna None si la cabeza de la lista no apunta a ningún nodo, indicando que la sala de espera está vacía.
        return self.cabeza is None
    
    # Busca un objeto paciente en la lista basándose en un valor específico y un criterio de comparación.
    def buscar_paciente(self, valor, criterio="nombre"):
        # Valida que la lista contenga información previo a realizar la búsqueda.
        if self.esta_vacia():
            # Cancela el proceso en caso de no haber elementos almacenados.
            return
        # Inicializa el puntero auxiliar en el primer elemento.
        actual = self.cabeza 
        
        # Sostiene la búsqueda mientras la posición actual sea válida.
        while actual is not None:
            # Compara el nombre actual con el valor de búsqueda si el criterio es "nombre".
            if criterio == "nombre" and actual.nombre_paciente.lower() == str(valor).lower():
                # Entrega el nodo que cumple con el criterio de texto.
                return actual
            # Compara el número de turno actual con el valor de búsqueda si el criterio es "turno".  
            elif criterio == "turno" and actual.numero_turno == valor:
                # Entrega el nodo que cumple con el criterio de número de turno.
                return actual
            # Salta al siguiente nodo disponible en el ordenamiento.
            actual = actual.sig
            
        # Retorna vacío si el bucle agota los elementos sin éxito.
        return None

    # Agrega un nuevo paciente al extremo final de la estructura doble.
    def llegada_paciente(self, turno, nombre, hora, motivo):
        # Llama a la función de búsqueda para verificar si el número de turno ya existe en la lista.
        turno_existente = self.buscar_paciente(turno, "turno")
        # Si se detecta un turno duplicado, se muestra un mensaje de error y se cancela la adición del nuevo paciente. 
        if turno_existente:
            print(f"Error: El turno {turno} ya está registrado para el paciente '{turno_existente.nombre_paciente}'. No se puede agregar el paciente '{nombre}'.")
            return
        # Instancia un nodo de la clase PacienteEspera con las variables dadas.
        nuevo_paciente = PacienteEspera(turno, nombre, hora, motivo) 
        
        # Evalúa si la fila se encuentra completamente desocupada.
        if self.cabeza is None:
            # Coloca al paciente recién creado como el origen de la lista.
            self.cabeza = nuevo_paciente 
        # Bloque de acción en caso de que ya existan nodos previos.
        else:
            # Define una referencia móvil posicionada en el primer nodo.
            actual = self.cabeza 
            
            # Recorre la secuencia mientras el enlace posterior sea válido.
            while actual.sig is not None:
                # Traslada la referencia móvil hacia el nodo subsiguiente.
                actual = actual.sig   
                
            # Enlaza la propiedad posterior del último nodo al nuevo paciente.
            actual.sig = nuevo_paciente   
            # Configura la propiedad anterior del nuevo paciente hacia el nodo penúltimo.
            nuevo_paciente.ant = actual   
            
        print(f"Paciente '{nombre}' registrado por {motivo}.")

    # Muestra el elemento que se encuentra inmediatamente después.
    def ver_siguiente(self, paciente_actual):
        # Valida que exista una referencia válida y que tenga un nodo posterior.
        if paciente_actual and paciente_actual.sig:
            print(f"Siguiente en espera: {paciente_actual.sig.nombre_paciente} - Motivo: {paciente_actual.sig.motivo_consulta}")
        # Acción si no se detecta ningún nodo a la derecha de la referencia.
        else:
            print(f"No hay más pacientes después de {paciente_actual.nombre_paciente}.")
    
    # Muestra el elemento que se encuentra inmediatamente antes.
    def ver_anterior(self, paciente_actual):
        # Comprueba la existencia del nodo de origen y su enlace previo.
        if paciente_actual and paciente_actual.ant:
            print(f"Anterior en espera: {paciente_actual.ant.nombre_paciente } - Motivo: {paciente_actual.ant.motivo_consulta}")
        # Acción si se detecta que el nodo es la cabeza de la estructura.
        else:
            print("Es el primer paciente.")
            
    # Reubica un nodo al inicio de la fila usando su número de turno.
    def prioridad_emergencia(self, turno):
        # Cancela el ordenamiento si la estructura está desprovista de nodos.
        if self.esta_vacia():
            print("La sala de espera está vacía. No se puede aplicar prioridad.")
            return
        
        # Comprueba si el turno indicado ya corresponde al primer nodo.
        if self.cabeza.numero_turno == turno:
                print("El/La paciente ya está en la primera posición de la fila.")
                return
        # Establece el puntero de exploración al comienzo de la lista.
        actual = self.cabeza
        # Analiza secuencialmente los nodos de la estructura doble.
        while actual is not None:
            # Verifica si localizó el identificador de turno buscado.
            if actual.numero_turno == turno:
                # Guarda temporalmente el puntero al vecino izquierdo.
                nodo_ant = actual.ant
                # Guarda temporalmente el puntero al vecino derecho.
                nodo_sig = actual.sig

                # Une el vecino izquierdo con el vecino derecho para aislar el nodo.
                nodo_ant.sig = nodo_sig
                # Valida que exista un nodo a la derecha antes de reconectar.
                if nodo_sig is not None:
                    # Une el vecino derecho con el izquierdo cerrando la brecha.
                    nodo_sig.ant = nodo_ant

                # Enlaza la propiedad siguiente del nodo a la cabeza actual de la lista.
                actual.sig = self.cabeza     
                # Modifica la propiedad anterior de la cabeza actual para apuntar al nodo que se moverá al frente.
                self.cabeza.ant = actual     
                # Declara el nodo actual como la nueva cabeza de la lista doble.
                self.cabeza = actual         
                # Rompe la referencia previa del nodo por ser ahora el primero.
                actual.ant = None            
                print(f"¡Prioridad aplicada! El/La paciente {actual.nombre_paciente} ahora es el primero.")
                # Concluye la función de forma inmediata tras el reordenamiento.
                return 
            # Avanza el puntero auxiliar al nodo de la derecha.
            actual = actual.sig
        print("Paciente con el turno especificado no encontrado. No se aplicó prioridad.")
        
    # Recorre toda la lista doble de principio a fin.
    def mostrar_lista_completa(self):
        # Cancela la lectura si la cola está vacía.
        if self.esta_vacia():
            print("La sala de espera está vacía.")
            return
        
        # Sitúa el lector auxiliar al inicio de la lista doble.
        actual = self.cabeza
        print("Lista completa de pacientes en espera:")
        # Mantiene la iteración activa hasta sobrepasar el último nodo.
        while actual is not None:
            print(f"Turno: {actual.numero_turno} - Nombre: {actual.nombre_paciente} - Hora Llegada: {actual.hora_llegada} - Motivo: {actual.motivo_consulta}")
            # Desplaza la lectura hacia el nodo derecho adyacente.
            actual = actual.sig
```

```python
# Define la función principal que administra el menú interactivo de la sala de espera.
def interfaz():
    # Inicializa una instancia limpia de la clase SalaEspera para gestionar la cola.
    sala = SalaEspera() 
    # Inicia un bucle de repetición indefinida para desplegar el panel de control.
    while True:
        print("\n CLÍNICA - SALA DE ESPERA")
        print("1. Registrar paciente")
        print("2. Buscar paciente por nombre en la fila")
        print("3. Ver siguiente en la fila") 
        print("4. Ver anterior en la fila") 
        print("5. Prioridad de Emergencia")
        print("6. Mostrar lista completa de espera")
        print("7. Salir")
        
        # Almacena el valor ingresado por el usuario para evaluar su decisión.
        opcion = input("Seleccione una opción: ")

        # Evalúa si la selección del menú coincide con la opción de registro.
        if opcion == "1":
            while True:
                # Abre un bloque de contención para interceptar fallas en la conversión de texto.
                try:
                    # Ingreso de turnos, nombres, horas y motivos de consulta.
                    turno = int(input("Número de turno: "))
                    nombre = input("Nombre del paciente: ").strip()
                    hora = int(input("Hora de llegada - formato 24h: "))
                    motivo = input("Motivo de la consulta: ").strip()
                    # Se validan los datos ingresados.
                    if turno > 0 and hora >= 0 and hora <= 23 and nombre and motivo:
                        # Invoca el método de inserción pasando las variables validadas.
                        sala.llegada_paciente(turno, nombre, hora, motivo)
                        break
                    # Bloque alterno si las variables de texto están vacías o los números son negativos.
                    else:
                        print("Datos no válidos. Intente de nuevo.")
                # Captura la inconsistencia de tipos si no se digitan valores numéricos válidos.
                except ValueError:
                    print("Asegúrese de ingresar correctamente los datos")

        # Valida si la opción escogida corresponde a buscar un elemento individual.
        elif opcion == "2":
            # Captura el parámetro de búsqueda textual eliminando los espacios sobrantes.
            n = input("Nombre del paciente a buscar: ").strip()
            # Asigna a una variable el resultado devuelto por el método de localización.
            encontrado = sala.buscar_paciente(n)
            
            # Evalúa si la referencia de búsqueda es equivalente a vacío.
            if encontrado is None:
                print("Paciente no encontrado en la sala de espera o la lista está vacía.")
            # Ejecuta la lectura de datos si se ubicó el nodo de manera exitosa.
            else:
                print(f"Datos: {encontrado.nombre_paciente} - Motivo: {encontrado.motivo_consulta} - Hora de llegada: {encontrado.hora_llegada}")

        # Comprueba si el usuario desea conocer al paciente ubicado a la derecha.
        elif opcion == "3":
            # Solicita el nombre que servirá como punto de origen para la lectura.
            n = input("¿Desde qué paciente quieres ver al siguiente?: ").strip()
            # Efectúa la búsqueda de la referencia dentro de la lista doble.
            encontrado = sala.buscar_paciente(n, "nombre")
            # Cancela el proceso de visualización si no existe el punto de referencia.
            if encontrado is None:
                print("Paciente no encontrado para mostrar el siguiente.")
            # Ejecuta la lectura del nodo adyacente derecho si la referencia es válida.
            else:
                # Envía el objeto localizado al visualizador del nodo posterior.
                sala.ver_siguiente(encontrado)
        
        # Determina si el comando solicitado requiere ver al paciente de la izquierda.
        elif opcion == "4":
            # Adquiere la identidad del paciente que actuará como eje de consulta.
            n = input("¿Desde qué paciente quieres ver al anterior?: ").strip()
            # Intenta rescatar el nodo coincidente dentro de la estructura.
            encontrado = sala.buscar_paciente(n)
            # Detiene el flujo de impresión si el nombre de partida no fue hallado.
            if encontrado is None:
                print("Paciente no encontrado para mostrar el anterior.")
            # Habilita la exploración hacia atrás al confirmar la existencia del nodo.
            else:
                # Llama al método de inspección anterior enviando el objeto de referencia.
                sala.ver_anterior(encontrado)
                
        # Verifica si se activó la opción para alterar la prioridad de la cola.
        elif opcion == "5":
            # Protege el proceso contra errores tipográficos en el campo de turno.
            try:
                # Transforma la entrada de la consola a un tipo de dato numérico entero.
                turno_prioridad = int(input("Ingrese el turno de prioridad: "))
                # Llama a la función encargada de reestructurar los punteros de la lista.
                sala.prioridad_emergencia(turno_prioridad)
            # Maneja la falla de conversión en caso de recibir caracteres alfabéticos.
            except ValueError:
                print("Error: Ingrese un número de turno entero válido.")

        # Valida si la opción elegida equivale a desplegar la cola completa.
        elif opcion == "6":
            # Invoca de forma directa al método de mapeo general de la estructura.
            sala.mostrar_lista_completa()
        # Evalúa si la instrucción digitada pide cerrar el programa.
        elif opcion == "7":
            print("Cerrando sistema de la sala de espera...")
            # Termina la iteración del ciclo para pausar el script.
            break
        
        # Recibe cualquier opción de menú que no esté contemplada numéricamente.
        else:
            print("Opción inválida.")

# Ejecuta la función principal para iniciar el sistema de gestión de la sala de espera.
interfaz()
```
