```python
# Clase Nodo representa cada canción en el árbol binario de búsqueda
class Nodo:
    def __init__(self, titulo, artista, genero, album, duracion, calificacion):
        # Datos del nodo (información de la canción)
        self.titulo = titulo
        self.artista = artista
        self.genero = genero
        self.album = album
        self.duracion = duracion # Esta es la clave principal para ordenar el BST
        self.likes = calificacion
        
        # Punteros a los subárboles izquierdo (menores) y derecho (mayores)
        self.izq = None
        self.der = None

# Clase ArbolBinarioBusqueda implementa un árbol binario de búsqueda (BST) para canciones
class ArbolBinarioBusqueda:
    def __init__(self):
        # Inicializo el árbol vacío, la raíz apunta a None
        self.raiz = None
    # Función para verificar si el árbol está vacío
    def esta_vacio(self):
        # Verifico si el árbol no tiene ningún nodo
        return self.raiz is None
    # Función para insertar un nuevo nodo en el árbol
    def insertar(self, titulo, artista, genero, album, duracion, calificacion):
        # Si está vacío, el nuevo nodo se convierte en la raíz
        if self.esta_vacio():
            # Instancio un nuevo nodo con los datos de la canción y lo asigno como raíz
            self.raiz = Nodo(titulo, artista, genero, album, duracion, calificacion)
        else:
            # Si ya hay nodos, inicio el recorrido recursivo desde la raíz
            self._insertar_recursivo(self.raiz, titulo, artista, genero, album, duracion, calificacion) 
    # Función recursiva para insertar un nodo en la posición correcta según la duración
    def _insertar_recursivo(self, actual, titulo, artista, genero, album, duracion, calificacion):
        # Si la duración es menor a la del nodo actual, me voy por la izquierda
        if duracion < actual.duracion:
            if actual.izq is None:
                # Si encuentro un espacio vacío a la izquierda, ahí coloco el nodo
                actual.izq = Nodo(titulo, artista, genero, album, duracion, calificacion)
            else:
                # Si está ocupado, sigo bajando recursivamente por la izquierda
                self._insertar_recursivo(actual.izq, titulo, artista, genero, album, duracion, calificacion)
        
        # Si es mayor o igual, me voy por la derecha
        elif duracion >= actual.duracion:
            if actual.der is None:
                # Si encuentro un espacio vacío a la derecha, lo inserto aquí
                actual.der = Nodo(titulo, artista, genero, album, duracion, calificacion)
            else:
                # Si está ocupado, sigo bajando recursivamente por la derecha
                self._insertar_recursivo(actual.der, titulo, artista, genero, album, duracion, calificacion)
    # Función para calcular el tiempo total de todas las canciones en el árbol, opcionalmente filtrando por álbum
    def calcular_tiempo_total(self, album_filtro=None):
        # Si el árbol está vacío, el tiempo total es 0
        if self.esta_vacio():
            return 0
        # Empiezo el cálculo sumando desde la raíz
        return self._calcular_tiempo_recursivo(self.raiz, album_filtro)
    # Función recursiva que suma las duraciones de los nodos, aplicando un filtro de álbum si se proporciona
    def _calcular_tiempo_recursivo(self, actual, album_filtro=None):
        # Caso base: Si llego a un nodo nulo, retorno 0
        if actual is None:
            return 0
        
        # Acumulo recursivamente los tiempos de los subárboles izquierdo y derecho mediante llamadas recursivas
        tiempo_izq = self._calcular_tiempo_recursivo(actual.izq, album_filtro)
        tiempo_der = self._calcular_tiempo_recursivo(actual.der, album_filtro)

        # Si no hay filtro o si el álbum del nodo coincide tomo su duración
        if album_filtro is None or normalizar_texto(actual.album) == normalizar_texto(album_filtro):
            # Si coincide, sumo la duración del nodo actual 
            duracion_actual = actual.duracion
        else:
            duracion_actual = 0 # Si no coincide, no suma nada
            
        # Retorno la suma de mis hijos más mi propia duración
        return tiempo_izq + tiempo_der + duracion_actual
    # Función para recomendar la canción más cercana a la duración deseada
    def recomendar_cancion_perfecta(self, segundos_disponibles):
        # Inicio la búsqueda O(log n) desde la raíz con valores iniciales
        return self._recomendar_recursivo(self.raiz, segundos_disponibles, None, float('inf'))
    # Función recursiva que busca la canción más cercana a la duración deseada
    def _recomendar_recursivo(self, actual, segundos_disponibles, mejor_nodo, menor_dif):
        if actual is None:
            # Si llegué al fondo, retorno el mejor candidato que encontré en el camino
            return mejor_nodo

        # Calculo la diferencia de tiempo entre el nodo actual y lo que busco
        dif_actual = abs(actual.duracion - segundos_disponibles)
        
        # Si esta diferencia es mejor que mi récord actual, actualizo al nuevo mejor candidato
        if dif_actual < menor_dif:
            mejor_nodo, menor_dif = actual, dif_actual

        # Si encontré una coincidencia exacta, corto la búsqueda de raíz
        if dif_actual == 0:
            return actual

        # Salto recursivo O(log n): Poda de ramas según si necesito más o menos tiempo
        if segundos_disponibles < actual.duracion:
            # Recursión hacia la izquierda si busco un tiempo menor
            return self._recomendar_recursivo(actual.izq, segundos_disponibles, mejor_nodo, menor_dif)
        else:
            # Recursión hacia la derecha si busco un tiempo mayor o igual
            return self._recomendar_recursivo(actual.der, segundos_disponibles, mejor_nodo, menor_dif)
    # Función para filtrar y eliminar canciones que sean más cortas que un umbral dado
    def filtrar_canciones_cortas(self, actual, duracion_minima):
        # Caso base: Si llego a un nodo nulo, retorno None para indicar que no hay más ramas
        if actual is None:
            return None
            
        # Poda estructural de abajo hacia arriba: Primero filtro los hijos izquierdo y derecho recursivamente
        actual.izq = self.filtrar_canciones_cortas(actual.izq, duracion_minima)
        actual.der = self.filtrar_canciones_cortas(actual.der, duracion_minima)

        # Si mi nodo es basura, devuelvo mi rama derecha para que tome mi lugar
        if actual.duracion < duracion_minima:
            return actual.der
            
        # Si sobrevivo, me devuelvo a mí mismo
        return actual
    
    # Función para eliminar un nodo específico por su duración, actualizando la raíz si es necesario
    def eliminar(self, duracion):
        # Actualizo la raíz por si cambia durante la eliminación
        self.raiz = self._eliminar_recursivo(self.raiz, duracion)
    # Función recursiva que busca y elimina un nodo por su duración, manejando los 3 casos de eliminación en BST
    def _eliminar_recursivo(self, actual, duracion):
        # Retorno None si llego a un nodo nulo, lo que indica que no encontré el valor
        if actual is None:
            return None
            
        # Navegación izquierda/derecha para encontrar el nodo a borrar
        # Si la duración buscada es menor, voy a la izquierda; si es mayor, voy a la derecha
        if duracion < actual.duracion:
            actual.izq = self._eliminar_recursivo(actual.izq, duracion)
        elif duracion > actual.duracion:
            actual.der = self._eliminar_recursivo(actual.der, duracion)
        else:
            # Si llego aquí, encontré el nodo a eliminar. Ahora manejo los 3 casos posibles:
            # Retorno el hijo derecho si no hay hijo izquierdo, o el hijo izquierdo si no hay hijo derecho
            if actual.izq is None:
                return actual.der
            if actual.der is None:
                return actual.izq
                
            # Si el nodo tiene ambos hijos, busco el sucesor (el nodo más pequeño del subárbol derecho)
            sucesor = self._encontrar_minimo(actual.der)
            
            # Copio todos los datos del sucesor al nodo actual (sin romper los punteros)
            actual.titulo, actual.artista, actual.genero, actual.album, actual.duracion, actual.likes = (
                sucesor.titulo, sucesor.artista, sucesor.genero, sucesor.album, sucesor.duracion, sucesor.likes
            )
            # Elimino físicamente al sucesor duplicado de su posición original
            actual.der = self._eliminar_recursivo(actual.der, sucesor.duracion)
        # Retorno el nodo actual (posiblemente modificado) para que el padre lo reconozca
        return actual
    # Función auxiliar para encontrar el nodo con la duración mínima en un subárbol
    def _encontrar_minimo(self, actual):
        # Bajo siempre por la izquierda para encontrar el valor más pequeño de un subárbol
        if actual.izq is None:
            # Si no hay más izquierda, este es el nodo mínimo
            return actual
        # Si hay un hijo izquierdo, sigo bajando recursivamente
        return self._encontrar_minimo(actual.izq)
    
    # Función para buscar un nodo por su título, devolviendo el nodo completo si se encuentra
    def buscar_titulo(self, titulo):
        titulo_norm = normalizar_texto(titulo)
        return self._buscar_titulo_recursivo(self.raiz, titulo_norm)
    # Función recursiva que busca un nodo por su título normalizado, recorriendo todo el árbol
    def _buscar_titulo_recursivo(self, actual, titulo_norm):
        # Caso base: Si llego a un nodo nulo, retorno None indicando que no encontré el título
        if actual is None:
            return None
        # Búsqueda exhaustiva (Pre-orden): Primero comparo el texto normalizado
        if normalizar_texto(actual.titulo) == titulo_norm:
            return actual
            
        # Si no, busco en la izquierda
        encontrado_izq = self._buscar_titulo_recursivo(actual.izq, titulo_norm)
        if encontrado_izq is not None:
            return encontrado_izq
            
        # Si tampoco, busco en la derecha
        return self._buscar_titulo_recursivo(actual.der, titulo_norm)
    # Función para buscar todos los nodos que pertenecen a un álbum específico, devolviendo un generador
    def album(self, album_buscar):
        album_norm = normalizar_texto(album_buscar)
        # Uso yield from para delegar la generación a la función recursiva
        yield from self._album_en_comun_recursivo(self.raiz, album_norm)
    # Función recursiva que recorre el árbol en orden y emite nodos que coinciden con el álbum buscado
    def _album_en_comun_recursivo(self, actual, album_norm):
        if actual is None:
            return
            
        # Visito todo el árbol emitiendo dinámicamente sin usar listas nativas
        yield from self._album_en_comun_recursivo(actual.izq, album_norm)
        # Si el álbum del nodo actual coincide con el buscado, lo emito al instante
        if normalizar_texto(actual.album) == album_norm: 
            yield actual
        # Continúo visitando el subárbol derecho
        yield from self._album_en_comun_recursivo(actual.der, album_norm)
    # Función para buscar canciones dentro de un rango de duración, devolviendo un generador
    def rango_duracion(self, duracion_minima, duracion_maxima):
        # Uso yield from para delegar la generación a la función recursiva que implementa la poda de ramas
        yield from self._busqueda_rango_recursivo(self.raiz, duracion_minima, duracion_maxima)
    # Función recursiva que implementa la búsqueda de rango con poda de ramas para eficiencia
    def _busqueda_rango_recursivo(self, actual, duracion_minima, duracion_maxima):
        # Caso base: Si llego a un nodo nulo, retorno sin emitir nada
        if actual is None:
            return
            
        # Poda izquierda, si mi nodo es mayor al mínimo, aún vale la pena revisar sus hijos menores
        if actual.duracion >= duracion_minima:
            yield from self._busqueda_rango_recursivo(actual.izq, duracion_minima, duracion_maxima)
            
        # Si el nodo entra en el rango, lo emito al instante
        if duracion_minima <= actual.duracion <= duracion_maxima:
            yield actual 
            
        # Poda derecha, si mi nodo es menor al máximo, vale la pena revisar sus hijos mayores
        if actual.duracion <= duracion_maxima:
            yield from self._busqueda_rango_recursivo(actual.der, duracion_minima, duracion_maxima)
    # Función para obtener todos los álbumes únicos del árbol, devolviendo una lista ordenada
    def obtener_todos_los_albumes(self):
        # Diccionario para agrupar claves únicas normalizadas y evitar duplicados de tipeo
        diccionario_albumes = {}
        # Recolecto todos los álbumes del árbol de manera recursiva
        self._recolectar_albumes_recursivo(self.raiz, diccionario_albumes)
        
        # Retorno una lista ordenada alfabéticamente solo para fines visuales en la interfaz
        return sorted(list(diccionario_albumes.values()))
    # Función recursiva que recorre el árbol y llena un diccionario con álbumes únicos
    def _recolectar_albumes_recursivo(self, actual, diccionario):
        if actual:
            # Normalizo el string para la comparación interna (ej. the wall)
            nombre_norm = normalizar_texto(actual.album)
            
            # Si es un álbum nuevo, lo guardo con su formato visual original (ej. The Wall)
            if nombre_norm not in diccionario:
                diccionario[nombre_norm] = actual.album
                
            # Sigo visitando el resto del árbol
            self._recolectar_albumes_recursivo(actual.izq, diccionario)
            self._recolectar_albumes_recursivo(actual.der, diccionario)
```

```python
# Configuración de la apariencia de la aplicación
ctk.set_appearance_mode("Dark")

BRICK_RED = "#9e1b1b"
BRICK_HOVER = "#7a1515"
WALL_WHITE = "#e0e0e0"
DARK_BG = "#0a0a0a"
# Clase principal de la aplicación de música
class AppMusica(ctk.CTk):
    # Ventana principal de la aplicación
    def __init__(self):
        super().__init__()
        self.title("Music App")
        self.geometry("1100x680")
        self.resizable(False, False)
        self.configure(fg_color=DARK_BG)

        self.arbol = ArbolBinarioBusqueda()
        for c in canciones_semilla:
            self.arbol.insertar(c["titulo"], c["artista"], c["genero"], c["album"], c["duracion"], c["calificacion"])
        
        self.reproduciendo = False

        self.sidebar = ctk.CTkFrame(self, width=280, corner_radius=0, fg_color="#141414")
        self.sidebar.pack(side="left", fill="y", padx=0, pady=0)
        self.sidebar.pack_propagate(False)
        
        self.main_container = ctk.CTkFrame(self, corner_radius=0, fg_color="transparent")
        self.main_container.pack(side="right", fill="both", expand=True, padx=20, pady=20)

        self.setup_sidebar()
        self.setup_player()
        self.setup_tabs()
    # Configuración de la barra lateral (sidebar)
    def setup_sidebar(self):
        lbl_logo = ctk.CTkLabel(self.sidebar, text="🧱 THE WALL MUSIC", font=ctk.CTkFont(size=22, weight="bold"), text_color=WALL_WHITE)
        lbl_logo.pack(padx=20, pady=(20, 5))
        lbl_sub = ctk.CTkLabel(self.sidebar, text="Ariel Jiménez", font=ctk.CTkFont(size=11, slant="italic"), text_color="gray")
        lbl_sub.pack(padx=20, pady=(0, 15))

        self.frame_metric = ctk.CTkFrame(self.sidebar, fg_color="#1c1c1c", border_color=BRICK_RED, border_width=1)
        self.frame_metric.pack(padx=15, pady=5, fill="x")
        
        self.lbl_metric_title = ctk.CTkLabel(self.frame_metric, text="DURACIÓN TOTAL PLAYLIST", font=ctk.CTkFont(size=10, weight="bold"), text_color=BRICK_RED)
        self.lbl_metric_title.pack(padx=10, pady=(5, 0), anchor="w")
        self.lbl_metric_val = ctk.CTkLabel(self.frame_metric, text="", font=ctk.CTkFont(size=18, weight="bold"), text_color=WALL_WHITE)
        self.lbl_metric_val.pack(padx=10, pady=(0, 5), anchor="w")
        
        btn_reset = ctk.CTkButton(self.sidebar, text="🔄 Ver Toda la Playlist", height=24, fg_color="#2b2b2b", hover_color="#3d3d3d", command=self.resetear_vista_playlist)
        btn_reset.pack(padx=15, pady=(0, 10), fill="x")

        ctk.CTkLabel(self.sidebar, text="📀 Explorar por Álbum", font=ctk.CTkFont(size=13, weight="bold"), text_color=WALL_WHITE).pack(padx=15, pady=(10, 0), anchor="w")
        
        self.scroll_albumes = ctk.CTkScrollableFrame(self.sidebar, height=130, fg_color="transparent")
        self.scroll_albumes.pack(padx=10, pady=5, fill="x")

        self.lbl_canciones_album = ctk.CTkLabel(self.sidebar, text="Pistas del Álbum:", font=ctk.CTkFont(size=12, weight="bold"), text_color="gray")
        self.lbl_canciones_album.pack(padx=15, pady=(5, 0), anchor="w")
        
        self.scroll_canciones = ctk.CTkScrollableFrame(self.sidebar, height=150, fg_color="#1c1c1c")
        self.scroll_canciones.pack(padx=15, pady=5, fill="both", expand=True)

        self.refrescar_sidebar_albumes()
        self.actualizar_tiempo_sidebar()
    # Configuración del reproductor de música
    def setup_player(self):
        self.player_frame = ctk.CTkFrame(self.main_container, height=130, border_width=2, border_color=BRICK_RED, fg_color="#181818")
        self.player_frame.pack(fill="x", padx=0, pady=(0, 15))
        self.player_frame.pack_propagate(False)

        self.lbl_status = ctk.CTkLabel(self.player_frame, text="⏸️ DETENIDO", font=ctk.CTkFont(size=11, weight="bold"), text_color=BRICK_RED)
        self.lbl_status.place(x=20, y=12)

        self.lbl_track_title = ctk.CTkLabel(self.player_frame, text="Ninguna pista seleccionada", font=ctk.CTkFont(size=18, weight="bold"), text_color=WALL_WHITE)
        self.lbl_track_title.place(x=20, y=32)

        self.lbl_track_meta = ctk.CTkLabel(self.player_frame, text="Artista: -- | Álbum: -- | Género: --", font=ctk.CTkFont(size=12), text_color="gray")
        self.lbl_track_meta.place(x=20, y=60)

        self.lbl_track_stars = ctk.CTkLabel(self.player_frame, text="Calificación: --", font=ctk.CTkFont(size=12), text_color="#FFD700")
        self.lbl_track_stars.place(x=20, y=82)

        self.btn_play = ctk.CTkButton(self.player_frame, text="▶️ Play", width=80, fg_color=BRICK_RED, hover_color=BRICK_HOVER, command=self.toggle_play)
        self.btn_play.place(x=650, y=45)
    # Configuración de las pestañas (tabs) de la aplicación
    def setup_tabs(self):
        self.tabview = ctk.CTkTabview(self.main_container, segmented_button_selected_color=BRICK_RED, segmented_button_selected_hover_color=BRICK_HOVER)
        self.tabview.pack(fill="both", expand=True)

        self.tab_sugerencia = self.tabview.add("💡 Sugerencia Inteligente")
        self.tab_limpieza = self.tabview.add("🧹 Limpieza de Basura")
        self.tab_rango = self.tabview.add("🚗 Modo Viaje")
        self.tab_buscar = self.tabview.add("🔍 Buscar Título")
        self.tab_anadir = self.tabview.add("➕ Añadir Canción")
        self.tab_eliminar = self.tabview.add("🗑️ Eliminar Canción")

        self.build_tab_sugerencia()
        self.build_tab_rango()
        self.build_tab_limpieza()
        self.build_tab_buscar()
        self.build_tab_anadir()
        self.build_tab_eliminar()
    # Construcción de la pestaña de sugerencias inteligentes
    def build_tab_sugerencia(self):
        ctk.CTkLabel(self.tab_sugerencia, text="Búsqueda de canciones", font=ctk.CTkFont(weight="bold")).pack(padx=20, pady=(15, 5), anchor="w")
        
        frame_input = ctk.CTkFrame(self.tab_sugerencia, fg_color="transparent")
        frame_input.pack(padx=20, pady=10, fill="x")

        ctk.CTkLabel(frame_input, text="Ingrese por teclado los segundos para buscar:").grid(row=0, column=0, padx=5)
        self.ent_segundos_sug = ctk.CTkEntry(frame_input, width=80)
        self.ent_segundos_sug.insert(0, "200")
        self.ent_segundos_sug.grid(row=0, column=1, padx=5)

        ctk.CTkLabel(self.tab_sugerencia, text="O ajusta mediante la barra:").pack(padx=20, pady=(10, 2), anchor="w")
        self.slider_seg = ctk.CTkSlider(self.tab_sugerencia, from_=60, to=480, width=400, button_color=BRICK_RED, button_hover_color=BRICK_HOVER, command=self.actualizar_input_desde_slider)
        self.slider_seg.pack(padx=20, pady=5, anchor="w")
        self.slider_seg.set(200)

        ctk.CTkButton(self.tab_sugerencia, text="✨ Encontrar Canción Perfecta", width=250, fg_color=BRICK_RED, hover_color=BRICK_HOVER, font=ctk.CTkFont(weight="bold"), command=self.ejecutar_recomendacion).pack(padx=20, pady=25)
    # Construcción de la pestaña de rango de canciones
    def build_tab_rango(self):
        ctk.CTkLabel(self.tab_rango, text="Seleccionar de canciones dentro de un rango de tiempo", font=ctk.CTkFont(weight="bold")).pack(padx=20, pady=10, anchor="w")
        frame_inputs = ctk.CTkFrame(self.tab_rango, fg_color="transparent")
        frame_inputs.pack(padx=20, pady=10, fill="x")

        ctk.CTkLabel(frame_inputs, text="Mínimo (seg):").grid(row=0, column=0, padx=5)
        self.ent_min = ctk.CTkEntry(frame_inputs, width=80)
        self.ent_min.insert(0, "180")
        self.ent_min.grid(row=0, column=1, padx=5)

        ctk.CTkLabel(frame_inputs, text="Máximo (seg):").grid(row=0, column=2, padx=25)
        self.ent_max = ctk.CTkEntry(frame_inputs, width=80)
        self.ent_max.insert(0, "260")
        self.ent_max.grid(row=0, column=3, padx=5)

        ctk.CTkButton(frame_inputs, text="🚗 Filtrar Rango", fg_color=BRICK_RED, hover_color=BRICK_HOVER, command=self.ejecutar_rango).grid(row=0, column=4, padx=30)

        self.txt_rango_output = ctk.CTkTextbox(self.tab_rango, height=180, font=ctk.CTkFont(size=12), fg_color="#1C1C1C", border_width=1, border_color="#333333")
        self.txt_rango_output.pack(padx=20, pady=10, fill="both", expand=True)
    # Construcción de la pestaña de limpieza de canciones cortas
    def build_tab_limpieza(self):
        ctk.CTkLabel(self.tab_limpieza, text="Eliminar automáticamente todas las canciones que duren menos de los segundos que usted indique", font=ctk.CTkFont(weight="bold")).pack(padx=20, pady=15, anchor="w")
        frame_del = ctk.CTkFrame(self.tab_limpieza, fg_color="transparent")
        frame_del.pack(padx=20, pady=10, fill="x")

        ctk.CTkLabel(frame_del, text="Elimine canciones menores a (segundos):").grid(row=0, column=0, padx=5)
        self.ent_umbral = ctk.CTkEntry(frame_del, width=80)
        self.ent_umbral.insert(0, "170")
        self.ent_umbral.grid(row=0, column=1, padx=10)

        ctk.CTkButton(frame_del, text="🧹 Ejecutar Limpieza de Audios Basura", fg_color="#8B0000", hover_color="#5c0000", command=self.ejecutar_poda).grid(row=0, column=2, padx=20)

        self.lbl_poda_stats = ctk.CTkLabel(self.tab_limpieza, text="Estado del árbol: Playlist original intacta.", font=ctk.CTkFont(size=13, slant="italic"), text_color="gray")
        self.lbl_poda_stats.pack(padx=20, pady=20, anchor="w")
    # Construcción de la pestaña de búsqueda de canciones por título
    def build_tab_buscar(self):
        ctk.CTkLabel(self.tab_buscar, text="Búsqueda de una canción por su nombre", font=ctk.CTkFont(weight="bold")).pack(padx=20, pady=15, anchor="w")
        frame_search = ctk.CTkFrame(self.tab_buscar, fg_color="transparent")
        frame_search.pack(padx=20, pady=10, fill="x")

        ctk.CTkLabel(frame_search, text="Título:").grid(row=0, column=0, padx=5)
        self.ent_titulo_buscar = ctk.CTkEntry(frame_search, width=250, placeholder_text="Ingrese el título de la canción a buscar")
        self.ent_titulo_buscar.grid(row=0, column=1, padx=10)

        ctk.CTkButton(frame_search, text="🔍 Buscar Pista", fg_color=BRICK_RED, hover_color=BRICK_HOVER, command=self.ejecutar_busqueda_titulo).grid(row=0, column=2, padx=15)
    # Construcción de la pestaña para añadir nuevas canciones
    def build_tab_anadir(self):
        ctk.CTkLabel(self.tab_anadir, text="Agregar Canción Nueva", font=ctk.CTkFont(weight="bold")).pack(padx=20, pady=(15, 5), anchor="w")
        ctk.CTkLabel(self.tab_anadir, text="Si no recuerda la información de la nueva canción, los campos vacíos se autocompletarán.", text_color="gray", font=ctk.CTkFont(size=11)).pack(padx=20, pady=(0, 15), anchor="w")

        f_form = ctk.CTkFrame(self.tab_anadir, fg_color="transparent")
        f_form.pack(padx=20, fill="x")

        ctk.CTkLabel(f_form, text="Título:").grid(row=0, column=0, padx=5, pady=5, sticky="e")
        self.ent_new_titulo = ctk.CTkEntry(f_form, width=150)
        self.ent_new_titulo.grid(row=0, column=1, padx=5, pady=5)

        ctk.CTkLabel(f_form, text="Artista:").grid(row=0, column=2, padx=5, pady=5, sticky="e")
        self.ent_new_artista = ctk.CTkEntry(f_form, width=150)
        self.ent_new_artista.grid(row=0, column=3, padx=5, pady=5)

        ctk.CTkLabel(f_form, text="Álbum:").grid(row=1, column=0, padx=5, pady=5, sticky="e")
        self.ent_new_album = ctk.CTkEntry(f_form, width=150, placeholder_text="Canción Nueva")
        self.ent_new_album.grid(row=1, column=1, padx=5, pady=5)

        ctk.CTkLabel(f_form, text="Género:").grid(row=1, column=2, padx=5, pady=5, sticky="e")
        self.ent_new_genero = ctk.CTkEntry(f_form, width=150)
        self.ent_new_genero.grid(row=1, column=3, padx=5, pady=5)

        ctk.CTkLabel(f_form, text="Duración (s):").grid(row=2, column=0, padx=5, pady=5, sticky="e")
        self.ent_new_dur = ctk.CTkEntry(f_form, width=80)
        self.ent_new_dur.grid(row=2, column=1, padx=5, pady=5, sticky="w")

        ctk.CTkLabel(f_form, text="Calificación (1-5):").grid(row=2, column=2, padx=5, pady=5, sticky="e")
        self.ent_new_cal = ctk.CTkEntry(f_form, width=80)
        self.ent_new_cal.grid(row=2, column=3, padx=5, pady=5, sticky="w")

        ctk.CTkButton(self.tab_anadir, text="➕ Insertar en Árbol", fg_color=BRICK_RED, hover_color=BRICK_HOVER, command=self.ejecutar_insercion).pack(padx=20, pady=20)
    # Construcción de la pestaña para eliminar canciones por duración
    def build_tab_eliminar(self):
        ctk.CTkLabel(self.tab_eliminar, text="Eliminar una canción exacta del árbol por su duración", 
                    font=ctk.CTkFont(weight="bold")).pack(padx=20, pady=15, anchor="w")
        ctk.CTkLabel(self.tab_eliminar, 
                    text="Ingrese la duración exacta de la canción que desea eliminar. Si existe una canción con esa duración, se eliminará de la playlist y se reestructurará automáticamente.",
                    text_color="gray", font=ctk.CTkFont(size=11), wraplength=650, justify="left").pack(padx=20, pady=(0, 15), anchor="w")

        frame_elim = ctk.CTkFrame(self.tab_eliminar, fg_color="transparent")
        frame_elim.pack(padx=20, pady=10, fill="x")

        ctk.CTkLabel(frame_elim, text="Duración exacta (segundos):").grid(row=0, column=0, padx=5)
        self.ent_dur_eliminar = ctk.CTkEntry(frame_elim, width=100)
        self.ent_dur_eliminar.grid(row=0, column=1, padx=10)

        ctk.CTkButton(frame_elim, text="🗑️ Eliminar Nodo", fg_color="#8B0000", hover_color="#5c0000", 
                    command=self.ejecutar_eliminacion).grid(row=0, column=2, padx=15)

        self.lbl_eliminar_stats = ctk.CTkLabel(self.tab_eliminar, text="", 
                                                font=ctk.CTkFont(size=13, slant="italic"), text_color="gray")
        self.lbl_eliminar_stats.pack(padx=20, pady=20, anchor="w")
    # Función para ejecutar la eliminación de una canción por duración
    def ejecutar_eliminacion(self):
        try:
            dur = int(self.ent_dur_eliminar.get())
        except ValueError:
            messagebox.showerror("Error", "Ingrese un valor numérico entero para la duración.")
            return

        objetivo = self.arbol.recomendar_cancion_perfecta(dur)
        if objetivo is None or objetivo.duracion != dur:
            messagebox.showwarning("No encontrado", f"No existe ninguna canción de exactamente {dur} segundos.")
            return

        titulo_eliminado = objetivo.titulo
        self.arbol.eliminar(dur)
        self.actualizar_tiempo_sidebar()
        self.refrescar_sidebar_albumes()

        self.lbl_eliminar_stats.configure(
            text=f"✅ Canción eliminada con éxito: '{titulo_eliminado}' ({dur}s).",
            text_color=BRICK_RED
        )
        messagebox.showinfo("Eliminación exitosa", f"Se eliminó la canción:\n'{titulo_eliminado}' ({dur}s)\nLa playlist y los álbumes se actualizaron.")
        self.ent_dur_eliminar.delete(0, "end")
    # Función para actualizar el valor del input de segundos según el slider
    def actualizar_input_desde_slider(self, val):
        self.ent_segundos_sug.delete(0, "end")
        self.ent_segundos_sug.insert(0, str(int(val)))
    # Función para actualizar la información de duración en la barra lateral (sidebar)
    def actualizar_tiempo_sidebar(self, album_filtro=None):
        total_seg = self.arbol.calcular_tiempo_total(album_filtro)
        if album_filtro:
            self.lbl_metric_title.configure(text=f"DURACIÓN: {album_filtro[:15].upper()}")
        else:
            self.lbl_metric_title.configure(text="DURACIÓN PLAYLIST TOTAL")
            
        self.lbl_metric_val.configure(text=f"{total_seg // 60}m {total_seg % 60}s ({total_seg}s)")
    # Función para resetear la vista de la playlist y mostrar todas las canciones
    def resetear_vista_playlist(self):
        self.actualizar_tiempo_sidebar(None)

        for widget in self.scroll_canciones.winfo_children():
            widget.destroy()
        self.lbl_canciones_album.configure(text="Pistas del Álbum:")
        self.lbl_status.configure(text="⏸️ MODO PLAYLIST GLOBAL", text_color="gray")
    # Función para refrescar la lista de álbumes en la barra lateral (sidebar)
    def refrescar_sidebar_albumes(self):
        for widget in self.scroll_albumes.winfo_children():
            widget.destroy()
            
        albumes = self.arbol.obtener_todos_los_albumes()
        for alb in albumes:
            frame_alb = ctk.CTkFrame(self.scroll_albumes, fg_color="transparent")
            frame_alb.pack(fill="x", pady=2)
            
            btn_play_alb = ctk.CTkButton(frame_alb, text="▶️", width=30, fg_color="#333333", hover_color=BRICK_RED, command=lambda a=alb: self.play_album_entero(a))
            btn_play_alb.pack(side="left", padx=(0,5))
            
            btn_alb = ctk.CTkButton(frame_alb, text=alb, fg_color="transparent", text_color=WALL_WHITE, hover_color="#2B2B2B", anchor="w", command=lambda a=alb: self.listar_canciones_album(a))
            btn_alb.pack(side="left", fill="x", expand=True)
    # Función para listar las canciones de un álbum específico en la barra lateral (sidebar)
    def listar_canciones_album(self, nombre_album):
        self.actualizar_tiempo_sidebar(nombre_album)
        self.lbl_canciones_album.configure(text=f"Pistas de: {nombre_album}")
        

        for widget in self.scroll_canciones.winfo_children():
            widget.destroy()
            
        encontrados = False
        for pista in self.arbol.album(nombre_album):
            encontrados = True
            f_pista = ctk.CTkFrame(self.scroll_canciones, fg_color="transparent")
            f_pista.pack(fill="x", pady=2)
            
            btn_play = ctk.CTkButton(f_pista, text="▶️", width=30, height=20, fg_color="#222222", hover_color=BRICK_RED, command=lambda p=pista: self.montar_pista_en_reproductor(p))
            btn_play.pack(side="left", padx=2)
            
            lbl_pista = ctk.CTkLabel(f_pista, text=f"{pista.titulo} ({pista.duracion}s)", font=ctk.CTkFont(size=11))
            lbl_pista.pack(side="left", padx=5)

        if not encontrados:
            ctk.CTkLabel(self.scroll_canciones, text="No hay pistas.", text_color="gray").pack()

    # Función para reproducir todas las canciones de un álbum específico
    def play_album_entero(self, nombre_album):
        self.reproduciendo = True
        self.lbl_status.configure(text=f"🎶 REPRODUCIENDO ÁLBUM", text_color=BRICK_RED)
        self.lbl_track_title.configure(text=nombre_album.upper())
        self.lbl_track_meta.configure(text="Reproducción de todas las pistas del álbum seleccionado.")
        self.lbl_track_stars.configure(text="")
        self.btn_play.configure(text="⏸️ Pausar")
        self.listar_canciones_album(nombre_album)
    # Función para montar la información de una pista en el reproductor
    def montar_pista_en_reproductor(self, nodo):
        if nodo is None: return
        self.nodo_actual_reproductor = nodo
        self.reproduciendo = True
        self.lbl_status.configure(text="🎶 REPRODUCIENDO CANCIÓN", text_color=BRICK_RED)
        self.lbl_track_title.configure(text=nodo.titulo)
        self.lbl_track_meta.configure(text=f"Artista: {nodo.artista}  |  Álbum: {nodo.album}  |  Género: {nodo.genero}  |  Tiempo: {nodo.duracion}s")
        self.lbl_track_stars.configure(text="⭐" * int(nodo.likes) + f" ({nodo.likes})")
        self.btn_play.configure(text="⏸️ Pausar")
    # Función para alternar entre reproducir y pausar la canción actual
    def toggle_play(self):
        if self.lbl_track_title.cget("text") == "Ninguna pista seleccionada":
            messagebox.showwarning("Reproductor", "Selecciona una canción o álbum del árbol primero.")
            return
        self.reproduciendo = not self.reproduciendo
        if self.reproduciendo:
            text_stat = "🎶 REPRODUCIENDO ÁLBUM" if "ÁLBUM" in self.lbl_status.cget("text") else "🎶 REPRODUCIENDO CANCIÓN"
            self.lbl_status.configure(text=text_stat, text_color=BRICK_RED)
            self.btn_play.configure(text="⏸️ Pausar")
        else:
            self.lbl_status.configure(text="⏸️ EN PAUSA", text_color="orange")
            self.btn_play.configure(text="▶️ Play")

    # Función para ejecutar la recomendación de canción perfecta según los segundos ingresados
    def ejecutar_recomendacion(self):
        try:
            segundos = int(self.ent_segundos_sug.get())
        except ValueError:
            messagebox.showerror("Error", "Ingrese un valor numérico para los segundos.")
            return

        ganadora = self.arbol.recomendar_cancion_perfecta(segundos)
        if ganadora:
            self.montar_pista_en_reproductor(ganadora)
            dif = abs(ganadora.duracion - segundos)
            if dif == 0:
                messagebox.showinfo("Búsqueda BST", f"¡Coincidencia Perfecta! Encontrada: '{ganadora.titulo}' de exactamente {segundos}s.")
            else:
                messagebox.showinfo("Búsqueda BST", f"Canción más cercana encontrada:\n\n'{ganadora.titulo}' ({ganadora.duracion}s)\nDiferencia absoluta de {dif} segundos.")
    # Función para ejecutar la búsqueda de canciones dentro de un rango de duración
    def ejecutar_rango(self):
        try:
            val_min = int(self.ent_min.get())
            val_max = int(self.ent_max.get())
        except ValueError:
            messagebox.showerror("Error", "Ingrese números enteros válidos en los rangos.")
            return

        self.txt_rango_output.delete("0.0", "end")
        self.txt_rango_output.insert("0.0", f"🚗 PLAYLIST DE VIAJE GENERADA (Rango: {val_min}s a {val_max}s)\n======================================================\n")
        
        encontradas = False
        tiempo_mix = 0
        
        for c in self.arbol.rango_duracion(val_min, val_max):
            encontradas = True
            tiempo_mix += c.duracion
            estrellas = "⭐" * int(c.likes)
            self.txt_rango_output.insert("end", f"🎵 Pista: {c.titulo:<22} | Artista: {c.artista:<18} | Duración: {c.duracion}s | {estrellas}\n")
        
        if not encontradas:
            self.txt_rango_output.insert("end", "No se encontraron canciones en este rango de tiempo.")
        else:
            minutos = tiempo_mix // 60
            segundos = tiempo_mix % 60
            self.reproduciendo = True
            self.lbl_status.configure(text="🎶 REPRODUCIENDO MIX MODO VIAJE", text_color="#9e1b1b") # BRICK_RED
            self.lbl_track_title.configure(text=f"Mix de Viaje ({val_min}s - {val_max}s)")
            self.lbl_track_meta.configure(text=f"Duración total del Mix: {minutos}m {segundos}s  |  Género: Variado")
            self.lbl_track_stars.configure(text="⭐⭐⭐⭐⭐ Playlist Automática")
            self.btn_play.configure(text="⏸️ Pausar")
    # Función para ejecutar la poda del árbol eliminando canciones cortas según el umbral ingresado
    def ejecutar_poda(self):
        try:
            umbral = int(self.ent_umbral.get())
        except ValueError:
            messagebox.showerror("Error", "Ingrese un umbral numérico válido.")
            return

        tiempo_antes = self.arbol.calcular_tiempo_total()
        self.arbol.raiz = self.arbol.filtrar_canciones_cortas(self.arbol.raiz, umbral)
        tiempo_despues = self.arbol.calcular_tiempo_total()

        self.actualizar_tiempo_sidebar()
        self.refrescar_sidebar_albumes()
        self.lbl_poda_stats.configure(
            text=f"✨ Limpieza ejecutada con éxito.\nTiempo total anterior: {tiempo_antes}s | Nuevo tiempo total: {tiempo_despues}s\nSe eliminaron todas las canciones menores a {umbral} segundos de la playlist.",
            text_color=BRICK_RED
        )
        messagebox.showinfo("Limpieza de Audios Basura", f"✨ Limpieza ejecutada con éxito.\nTiempo total anterior: {tiempo_antes}s | Nuevo tiempo total: {tiempo_despues}s\nSe eliminaron todas las canciones menores a {umbral} segundos de la playlist.")

    # Función para ejecutar la búsqueda de una canción por su título
    def ejecutar_busqueda_titulo(self):
        buscado = self.ent_titulo_buscar.get().strip()
        if not buscado:
            messagebox.showwarning("Búsqueda", "Escribe un título para iniciar el recorrido de búsqueda.")
            return
        
        res = self.arbol.buscar_titulo(buscado)
        if res:
            self.montar_pista_en_reproductor(res)
            messagebox.showinfo("Éxito", f"¡Pista '{res.titulo}' localizada!\nEnviada al reproductor.")
        else:
            messagebox.showerror("No encontrado", f"La canción '{buscado}' no existe en la playlist.\n(Pudo haber sido eliminada por la limpieza de audios basura).")
    # Función para ejecutar la inserción de una nueva canción en el árbol
    def ejecutar_insercion(self):
        tit = (self.ent_new_titulo.get().strip() or "Desconocido").title()
        art = (self.ent_new_artista.get().strip() or "Desconocido").title()
        gen = (self.ent_new_genero.get().strip() or "Desconocido").title()
        alb = (self.ent_new_album.get().strip() or "Canción Nueva").title()
        
        try:
            dur = int(self.ent_new_dur.get()) if self.ent_new_dur.get().strip() else random.randint(150, 320)
            cal = float(self.ent_new_cal.get()) if self.ent_new_cal.get().strip() else round(random.uniform(3.0, 5.0), 1)
        except ValueError:
            messagebox.showerror("Error", "Asegúrate de que la duración y calificación sean números.")
            return

        self.arbol.insertar(tit, art, gen, alb, dur, cal)
        self.actualizar_tiempo_sidebar()
        self.refrescar_sidebar_albumes()
        
        self.ent_new_titulo.delete(0, "end")
        self.ent_new_artista.delete(0, "end")
        self.ent_new_album.delete(0, "end")
        self.ent_new_genero.delete(0, "end")
        self.ent_new_dur.delete(0, "end")
        self.ent_new_cal.delete(0, "end")
        
        messagebox.showinfo("Inserción Exitosa", f"Se ha insertado la canción:\n'{tit}' ({dur}s)\n¡La playlist y los álbumes se han actualizado!")
```
