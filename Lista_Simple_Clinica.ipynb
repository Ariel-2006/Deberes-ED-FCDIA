```python
# Define la estructura de cada nodo que representa una consulta médica.
class AtencionMedica:
    # Constructor de la clase que inicializa las propiedades del nodo.
    def __init__(self, id_registro: int, nombre_paciente: str, fecha: str, monto_consulta: float):
        # Define el identificador único del registro médico.
        self.id_registro = id_registro
        # Guarda el nombre completo del paciente.
        self.nombre_paciente = nombre_paciente
        # Registra la fecha de la consulta médica.
        self.fecha = fecha
        # Almacena el costo de la consulta.
        self.monto_consulta = monto_consulta
        # Inicializa el puntero de enlace al siguiente nodo como nulo.
        self.sig = None

# Define la estructura de la lista enlazada para gestionar las consultas.
class HistorialMedico:  
    # Constructor de la lista que inicializa la estructura sin elementos.
    def __init__(self):
        # Establece el inicio de la lista enlazada como vacío.
        self.cabeza = None   

    # Método para comprobar si la lista carece de registros.
    def esta_vacio(self):
        # Retorna verdadero si la lista no tiene ningún nodo inicial.
        return self.cabeza is None

    # Método para localizar un registro mediante el nombre del paciente o el ID de registro.
    def buscar_registro(self, valor, criterio="nombre"):
        # Verifica si la lista carece de elementos antes de iniciar la búsqueda.
        if self.esta_vacio():
            print("El historial médico está vacío.")
            # Interrumpe la función al no haber datos que procesar.
            return
        
        # Posiciona el apuntador auxiliar al inicio de la lista.
        actual = self.cabeza 
        
        # Mantiene el ciclo activo mientras el apuntador haga referencia a un nodo válido.
        while actual is not None:
            # Compara el nombre del nodo con el valor buscado si el criterio es "nombre".
            if criterio == "nombre" and actual.nombre_paciente.lower() == str(valor).lower():
                # Devuelve el nodo que coincide con el nombre del paciente buscado.
                return actual
            # Compara el ID del nodo con el valor buscado si el criterio es "id".
            elif criterio == "id" and actual.id_registro == valor:
                # Retorna el nodo que coincide con el ID de registro buscado.
                return actual  
            
            # Mueve el apuntador hacia el siguiente nodo de la lista.
            actual = actual.sig 
            
        # Retorna un valor nulo si termina el recorrido sin coincidencias.
        return None
    
    # Método para insertar un nuevo nodo al final de la lista.
    def agregar_registro(self, id_registro, nombre_paciente, fecha, monto_consulta):
        # Llama a la función de búsqueda para verificar si el ID de registro ya existe en la lista.
        registro_existente = self.buscar_registro(id_registro, "id")
        # Si se detecta un ID de registro duplicado, se muestra un mensaje de error y se cancela la adición del nuevo registro.
        if registro_existente:
            print(f"Error: El ID de registro {id_registro} ya existe para el paciente '{registro_existente.nombre_paciente}'. No se puede agregar el nuevo registro.")
            return
        # Crea una nueva instancia de la clase AtencionMedica.
        nuevo_registro = AtencionMedica(id_registro, nombre_paciente, fecha, monto_consulta) 
        
        # Evalúa si la lista se encuentra actualmente sin elementos.
        if self.cabeza is None:
            # Asigna el nuevo registro como el primer elemento de la lista.
            self.cabeza = nuevo_registro
        # Bloque de ejecución si la lista ya contiene elementos almacenados.
        else:
            # Crea un apuntador auxiliar en el primer nodo para recorrer la lista.
            actual = self.cabeza 
            
            # Recorre los nodos mientras exista un nodo posterior enlazado.
            while actual.sig is not None:
                # Desplaza el apuntador auxiliar hacia el siguiente nodo.
                actual = actual.sig 
            
            # Conecta el último nodo de la lista con el nuevo registro.
            actual.sig = nuevo_registro 
        
        print(f"Registro {id_registro} de '{nombre_paciente}' añadido al historial.")

    # Método para recorrer de principio a fin todos los nodos.
    def mostrar_historial_completo(self):
        # Comprueba si la lista no contiene registros para mostrar.
        if self.esta_vacio():
            print("El historial médico está vacío.")
            # Finaliza la ejecución de la función.
            return
            
        # Coloca el apuntador de lectura en el nodo inicial.
        actual = self.cabeza

        # Recorre la lista secuencialmente de principio a fin.
        while actual is not None:
            print(f"ID Registro: {actual.id_registro} | Nombre: {actual.nombre_paciente} | Fecha: {actual.fecha} | Monto Consulta: ${actual.monto_consulta:.2f}")
            
            # Pasa al siguiente nodo de la lista enlazada.
            actual = actual.sig

    # Método para filtrar consultas que superen un costo financiero determinado.
    def filtrar_por_monto(self, valor_minimo):
        # Valida si la estructura está vacía para cancelar el filtro.
        if self.esta_vacio():
            print("El historial médico está vacío.")
            # Termina la función al no haber datos que procesar para el filtro.
            return
            
        # Apunta al primer nodo para iniciar la evaluación del filtro.
        actual = self.cabeza
        print(f"Montos mayores a {valor_minimo}")
        
        # Itera a través de todos los nodos de la estructura.
        while actual is not None:
            # Compara si el monto del registro actual supera el límite establecido.
            if actual.monto_consulta > valor_minimo:
                print(f"ID Registro: {actual.id_registro} | Nombre: {actual.nombre_paciente} | Fecha: {actual.fecha} | Monto Consulta: ${actual.monto_consulta:.2f}")
            
            # Avanza el apuntador al siguiente elemento disponible.
            actual = actual.sig
```

```python
# Define la función principal que gestiona el menú interactivo del sistema.
def interfaz():
    # Inicializa una instancia de la clase para almacenar los registros médicos.
    historial = HistorialMedico() 
    
    # Inicia un bucle para mantener el menú operativo de forma continua.
    while True:
        print("\nCLÍNICA - HISTORIAL MÉDICO")
        print("1. Agregar registro de atención")
        print("2. Buscar historial de paciente")
        print("3. Filtrar consultas por monto mínimo")
        print("4. Mostrar historial completo")
        print("5. Salir")
        
        # Captura la entrada de texto del usuario en la consola.
        opcion = input("Seleccione una opción: ")

        # Evalúa si el usuario seleccionó la primera opción del menú.
        if opcion == "1":
            # Inicia un ciclo de captura que se repetirá hasta recolectar información válida.
            while True:
                # Abre un bloque de contención interno para manejar fallas de formato en cada intento.
                try:
                    # Ingreso de id de registro, nombre del paciente, fecha de consulta y monto de la atención.
                    id_reg = int(input("ID de Registro: "))
                    nombre = input("Nombre del paciente: ").strip()
                    fecha = input("Fecha (ej. 12/05/2026): ").strip()
                    monto = float(input("Monto de la consulta: $"))
                    
                    # Valida que los datos que deben ser numéricos sean positivos y que los campos de texto no estén vacíos.
                    if id_reg > 0 and monto >= 0 and nombre and fecha:
                        # Invoca al método de la clase para guardar el nuevo registro en la lista.
                        historial.agregar_registro(id_reg, nombre, fecha, monto)
                        # Rompe el bucle de captura tras almacenar exitosamente la información.
                        break
                    # Bloque alterno si las variables de texto están vacías o los números son negativos.
                    else:
                        print("Datos no válidos. Intente de nuevo.")
                # Captura la inconsistencia si se digita letras en campos numéricos.
                except ValueError:
                    print("Asegúrese de ingresar correctamente los datos.")

        # Verifica si la opción ingresada corresponde a la búsqueda de un paciente.
        elif opcion == "2":
            # Captura el nombre a buscar y elimina los espacios innecesarios.
            n = input("Nombre del paciente a buscar: ").strip()
            # Ejecuta el método de búsqueda y almacena el nodo recuperado.
            encontrado = historial.buscar_registro(n, "nombre")
            
            # Evalúa si la búsqueda regresó un valor nulo.
            if encontrado is None:
                print("No se encontró ningún registro para ese paciente.")
            # Bloque de ejecución secundario en caso de que el registro sí exista.
            else:
                print(f"Historial encontrado: ID {encontrado.id_registro} - Paciente: {encontrado.nombre_paciente} - Fecha: {encontrado.fecha} - Monto: ${encontrado.monto_consulta}")

        # Filtro de consultas por monto mínimo.
        elif opcion == "3":
            # Inicia un bloque de protección contra datos de tipo texto inválidos.
            try:
                # Convierte el valor mínimo ingresado a un dato de tipo decimal.
                valor_min = float(input("Ingrese valor mínimo: "))
                # Llama al método de filtrado pasando el umbral numérico obtenido.
                historial.filtrar_por_monto(valor_min)
            # Controla la falla generada por un formato numérico erróneo.
            except ValueError:
                print("Ingrese un número válido para el monto.")

        # Valida si la opción digitada requiere listar toda la información.
        elif opcion == "4":
            print("Mostrando historial completo:")
            # Ejecuta el método encargado de imprimir la lista enlazada completa.
            historial.mostrar_historial_completo()

        # Determina si el usuario ha solicitado abandonar el programa.
        elif opcion == "5":
            print("Saliendo del sistema de historial médico...")
            # Rompe el bucle iterativo para concluir la ejecución del script.
            break
        
        # Captura cualquier otra entrada que no coincida con los números del menú.
        else:
            print("Opción inválida.")

# Llama a la función principal para iniciar el sistema de gestión del historial médico.
interfaz()
```
